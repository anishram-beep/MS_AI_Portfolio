```python
import yfinance as yf
import pandas as pd
import time
```

```python
df = pd.read_csv(r'C:\Users\phata\OneDrive - George Mason University - O365 Production\AIT626\05. Projects\Final Project\Earnings2Insights\companylist.csv', parse_dates=['CloseDay', 'CloseDay+1', 'CloseDay+7', 'CloseDay+28'])
```

```python
for col in ['CloseDay', 'CloseDay+1', 'CloseDay+7', 'CloseDay+28']:
    df[col] = pd.to_datetime(df[col]).dt.strftime('%Y-%m-%d')

df.head()
```

```python
def get_price_on_or_before(ticker, date):
    dat = yf.Ticker(ticker)
    hist = dat.history(period="7d", end=date + pd.Timedelta(days=1), auto_adjust=False)
    if not hist.empty:
        return hist["Close"].iloc[-1].item()
    else:
        return None
```

```python
for col in ['CloseDay', 'CloseDay+1', 'CloseDay+7', 'CloseDay+28']:
    price_col = col + "_Price"
    df[price_col] = None

    for i, row in df.iterrows():
        ticker = row['Ticker']
        date = row[col]
        if pd.notna(date):
            try:
                df.at[i, price_col] = get_price_on_or_before(ticker, date)
                print(f"✅ {ticker} {col}: {date.date()} OK")
            except Exception as e:
                print(f"⚠️ {ticker} {col} failed: {e}")
                df.at[i, price_col] = None
            time.sleep(0.5)  # polite pause for Yahoo rate limits
```

```python
df.to_csv(r'C:\Users\phata\OneDrive - George Mason University - O365 Production\AIT626\05. Projects\Final Project\Earnings2Insights\companyPrices.csv', index=False)
print("\n💾 Saved to stock_prices_event_window.csv")
```
